```python
import pandas as pd 
df1=pd.read_csv("C:\\Users\Rahul\\Downloads\\df1_c.csv")
df2=pd.read_csv("C:\\Users\\Rahul\\Downloads\\df2_c.csv")
merged_df=pd.merge(df1,df2,on='customer_id')
```

```python
merged_df.duplicated().sum()
merged_df.drop_duplicates()
```

```python
def fill_nulls(merged_df):
    for col in merged_df.columns:
        if col=='amount':
            merged_df[col]=merged_df[col].fillna(merged_df['amount'].mean())
        elif merged_df[col].dtype=='object':
            merged_df[col]=merged_df[col].fillna(f'unknown_{col}')
        else :
            merged_df[col]=merged_df[col].fillna(0) 
    return merged_df

cleaned_data=fill_nulls(merged_df)

cleaned_data

                                                
          
    

```
